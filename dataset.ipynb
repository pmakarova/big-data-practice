```python
import pandas as pd
```

```python
df = pd.read_parquet('yellow_tripdata_2026-01.parquet')
df.head()
```

```python
df.shape[0]
```

```python
df = df[(df['tpep_pickup_datetime'] >= '2026-01-01') & (df['tpep_pickup_datetime'] <= '2026-01-31 23:59:59')]
df
```

```python
df = df[df['total_amount'] >= 0]
df
```

```python
df = df[df['passenger_count'] > 0]
df = df.dropna(subset=['passenger_count'])
df
```

```python
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
duration_hours = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 3600
duration_hours
```

```python
df = df[(duration_hours > 0) & (duration_hours < 5)] 
df
```
